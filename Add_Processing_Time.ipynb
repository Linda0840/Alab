```python
import numpy as np
import csv
import json
import pandas as pd
from pandas import json_normalize
import matplotlib.pyplot as plt
```

```python
training = pd.read_csv(
    "training.csv",
    sep="|",
    engine="python",
    encoding="utf-8-sig",
    quoting=csv.QUOTE_NONE,   # ‚Üê treat " as a normal character
    escapechar="\\",          # backslash-escaped quotes like \" stay intact
    on_bad_lines="skip"       # or "warn"
)

print(training.shape)
```

```python
target_train = pd.read_csv('target_train.csv')
print(target_train.shape)
target_test = pd.read_csv('target_test.csv')
print(target_test.shape)
```

```python
missing_col = ["APPROVAL_PROCESSED_TIME_INFO"]

target_train = target_train.merge(
    training[["APPROVAL_ID_INFO"] + missing_col],
    on="APPROVAL_ID_INFO",
    how="left"
)

target_test = target_test.merge(
    training[["APPROVAL_ID_INFO"] + missing_col],
    on="APPROVAL_ID_INFO",
    how="left"
)
```

```python
# should automatically overwrite
# check if now have 16 columns

target_train.to_csv("target_train.csv", index=False)
target_test.to_csv("target_test.csv", index=False)
```
